Create violin plots of expression of Nfi transcription factors
Nfia,Nfib,Nfic,Nfix in Leydig Cells (Ernst et al. 2019)

```r
library(ggplot2)
library(Matrix)
```

```r
# Call plotViolinExpression from a file
source("functions/violin_plot_gene_exp.R")
```

Data generated by import_prepare_ErnstData_Scran.ipynb

```r
# Load Leydig P15 experiment file


load("data/ernst/LeydigP15.Rdata")
LeydigP15
```

```r
NfiGenees = c("Nfia","Nfib","Nfic","Nfix")
plotViolinExpression(LeydigP15,"LeydigP15",NfiGenees,zeros=TRUE,ylim=TRUE,ylim0=0,ylim1=4.5)
```
