Create violin plots of expression of Nfi transcription factors
Nfia,Nfib,Nfic,Nfix in Spermatogonia Cells

```r
library(ggplot2)
library(Matrix)
```

```r
# Call plotViolinExpression from a file
source("functions/violin_plot_gene_exp.R")

```

Data generated by  import_prepare_ErnstData_Scran.ipynb

```r
# Load spermatogonia sce data SpermatogoniaP15


load("data/ernst/SpermatogoniaP15.Rdata")

SpermatogoniaP15
```

```r
######### Plot function P15
NfiGenees = c("Nfia","Nfib","Nfic","Nfix")
plotViolinExpression(SpermatogoniaP15,"SpermatogoniaP15",NfiGenees, zeros=TRUE,ylim=TRUE,ylim0=0,ylim1=4.5)
```
