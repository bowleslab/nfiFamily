Create violin plots of expression of Nfi transcription factors
Nfia,Nfib,Nfic,Nfix in Spermatocytes Cells (Ernst et al. 2019)

```r
library(ggplot2)
library(Matrix)
```

```r
# Call plotViolinExpression from a file
source("functions/violin_plot_gene_exp.R")
```

Data generated by  import_prepare_ErnstData_Scran.ipynb

```r
# Load spermatocyte data


load("data/ernst/SpermatocytesP15.Rdata")

SpermatocytesP15
```

```r
NfiGenees = c("Nfia","Nfib","Nfic","Nfix")
plotViolinExpression(SpermatocytesP15,"SpermatocytesP15",NfiGenees,zeros=TRUE,ylim=TRUE,ylim0=0,ylim1=4.5)
```
