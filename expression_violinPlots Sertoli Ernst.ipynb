Create violin plots of expression of Nfi transcription factors
Nfia,Nfib,Nfic,Nfix in Sertoli Cells

```r
library(ggplot2)
library(Matrix)
```

```r
# Call plotViolinExpression from a file
source("functions/violin_plot_gene_exp.R")
```

Data generated by import_prepare_ErnstData_Scran.ipynb

```r
# Load Sertoli P15 experiment file


load("data/ernst/SertoliP15.Rdata")
SertoliP15
```

```r
NfiGenees = c("Nfia","Nfib","Nfic","Nfix")
plotViolinExpression(SertoliP15,"SertoliP15",NfiGenees,zeros=TRUE,ylim=TRUE,ylim0=0,ylim1=4.5)
```
